```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img  = cv2.imread('week7_homework.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
kernel = np.ones((9,9),np.float32) / 81

dst = cv2.filter2D(img, -1, kernel)
clo = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
blur = cv2.GaussianBlur(img, (9,9), 0)
median = cv2.medianBlur(img, 11)

plt.figure(figsize=(20,350))
plt.subplot(151), plt.imshow(img), plt.title('Original')
plt.subplot(152), plt.imshow(dst), plt.title('filtered'), plt.axis('off')
plt.subplot(153), plt.imshow(clo), plt.title('closed'), plt.axis('off')
plt.subplot(154), plt.imshow(blur), plt.title('GaussianBlur'), plt.axis('off')
plt.subplot(155), plt.imshow(median), plt.title('median'), plt.axis('off')
plt.show()
```
